```python
import pandas as pd,matplotlib.pyplot as plt,seaborn as sns,numpy as np
from matplotlib.pyplot import xticks
import os



#Loading the dataset to Categorize and analyze.
df = pd.read_csv("clean_titanic_data.csv")
```

```python
#mapping the Survival column from zeros and ones to Yess and No
map_survived = {0:"NO",1:"YES"}
df["Survived_Category"] = df["Survived"].map(map_survived)
```

```python
# mapping the passenger column entries from 1,2,3 to First,Second and Third respectively as classes.
map_pclass = {1:"First",2:"Second",3:"Third"}
df["Pclass_Category"] = df["Pclass"].map(map_pclass)
```

```python
#Creating a new SibSp_category the entries of sibsp column to it categorical corresponding.
map_SibSp = {0:"Alone",1:"With 1 Sib/Sp",2:"Small FM",3:"Medium FM",4:"Large FM",5:"V Large FM",8:"Extended FM"}
df["SibSp_Category"] = df["SibSp"].map(map_SibSp)
```

```python
#creating a new column of Parch_Category mapping numerical Parch to categorical.
map_parch = {0:"Alone",1:"With 1 Par/Chld",2:"Small FM",3:"Medium FM",4:"Large FM",5:"V Large FM",6:"Extended FM"}
df["Parch_Category"] = df["Parch"].map(map_parch)
```

```python
#visualizing the categorical data,Survival by parch category,survival by SibsP ,survival by sex,correlation matrix heatmap to view relationship btw columns.
fig,axes = plt.subplots(2,2,figsize =(16,18))
#Survival Rate vs Siblings/Spouse category
sns.countplot(x = "SibSp_Category",hue = "Survived",data = df,ax = axes[0,0])
axes[0,0].set_title("Survived vs SibSp")
#Survival Rate vs Parch_category
sns.countplot(x = "Parch_Category",hue = "Survived",data = df,ax =axes[0,1])
axes[0,1].set_title("Survived vs Parch")
#correlation heat map to view relationship btw colums
sns.heatmap(df.corr(numeric_only  = True),annot = True,cmap = "coolwarm",ax = axes[1,0])
axes[1,0].set_title("Correlation Matrix")
#survival count vs gender type
sns.countplot(x = "Sex",hue = "Survived",data = df,ax = axes[1,1])
plt.savefig("Figures/chart.jpg")
plt.show()

```
