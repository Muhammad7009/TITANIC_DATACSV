```python
import pandas as pd,matplotlib.pyplot as plt,seaborn as sns
df = pd.read_csv("clean_titanic_data.csv")
```

```python
group_survive_by_pclass = df.groupby("Pclass")["Survived"].sum().reset_index()
group_survive_by_pclass["Survived"] = (group_survive_by_pclass["Survived"]/group_survive_by_pclass["Survived"].sum()).mul(100)
```

```python
plt.figure()
plt.pie(group_survive_by_pclass["Survived"],labels = group_survive_by_pclass["Pclass"], shadow = True, colors = sns.color_palette("hls", 10),autopct = "%1.1f%%",explode = [0.2,0,0])

plt.show()
plt.savefig("chart.png")
```

```python
group_survive_gender = df.groupby("Sex")["Survived"].mean().mul(100).reset_index()
group_survive_gender
```

```python
plt.figure()
plt.pie(group_survive_gender["Survived"],labels = group_survive_gender["Sex"],autopct = "%1.1f%%",colors = sns.color_palette("hls", 10),explode = [0.2,0])
plt.show()
```
